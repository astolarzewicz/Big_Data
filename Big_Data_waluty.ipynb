```python
import requests
import matplotlib.pyplot as plt
import pandas as pd

# ------- Dolar --------
USD = requests.get("https://api.nbp.pl/api/exchangerates/rates/a/usd/2020-01-01/2020-07-18/")
usd_frame = pd.read_json(USD.content)
usd_rates = usd_frame.rates
usd_list=[index['mid'] for index in usd_rates]

USD19 = requests.get("https://api.nbp.pl/api/exchangerates/rates/a/usd/2019-01-01/2019-07-18/")
usd19_frame = pd.read_json(USD19.content)
usd19_rates = usd19_frame.rates
usd19_list=[index['mid'] for index in usd19_rates]

plt.plot(usd_list, '-g', usd19_list, '-b')
plt.xticks(rotation = 45)
plt.title("Kurs dolara w analogicznych okresach 2019 i 2020 r.")
plt.xlabel("Liczba dni od rozpoczÄ™cia roku")
plt.ylabel("Cena [PLN]")
plt.legend(["USD 2020", "USD 2019"])
plt.grid()

plt.show()
```
